### columns:
1 and 2 time (Time is always in chunks, with features calculated over 4 s windows.), \
3-131 features, \
132-134 triggers. 

Column 1: Time start \
Column 2: Time end \
\
Then the features are in 6 types: \
Column 3-6 EOG (eye movements) \
Column 7-8 ACC (accelerometer, head movement) \
Column 9-13 EDA (skin conductance) \
Column 14-16 EMG (face muscle activity) \
Column 17-72 EEG (eeg, in 4 wavebands, eeg theta, alpha, beta, and gamma) \
Column 73-132 EEG connectivity (in 4 wavebands) \
\
Column 133: Trigger values \
Column 134 : Actual time \
Column 135 : Duration (ignored) \
\
Notes on Column 133: \
Sometimes multiple triggers occur in the same window, in which case they are separated by a colon (:) \
the trigger values are too high by 33024 - i.e. subtract from each trigger 33024 to get the right identity \
Trigger to identity conversion. These triggers identify the START of a task: \
223: Training, negative valence \
224: Training, positive valence \
233: Testing, negative valence \
234: Testing, positive valence \
Trigger 7 identifies the END of a task \

```python
# Labeling the data
train_neg = '33247'
train_pos = '33248'
test_neg = '33257'
test_pos = '33258'
end = '33031'

# non null data
non_null_data = data[data['Event'].notnull()]
```

```python
# cast into string
non_null_data['Event'] = non_null_data['Event'].astype(str)
non_null_data['Event.1'] = non_null_data['Event.1'].astype(str)
```

```python
non_null_data.head(5)
```

```python
# iterating all to get start time and end time
train_neg_list = []
train_pos_list = []
test_neg_list = []
test_pos_list = []

tmp_baseline = 0
tmp_start_time = 0
tmp_end_time = 0
# 0 = not occuring, 1 = train_neg, 2 = train_pos, 3 = test_neg, 4 = test_pos
types = 0

for index, row in non_null_data.iterrows():
    event = row['Event'].split(':')
    time = row['Event.1'].split(':')
    
    if types == 0:
        idx = -1
        # find the start time
        if train_neg in event:
            idx = event.index(train_neg)
            types = 1
        if train_pos in event:
            idx = event.index(train_pos)
            types = 2
        if test_neg in event:
            idx = event.index(test_neg)
            types = 3
        if test_pos in event:
            idx = event.index(test_pos)
            types = 4
        
        if idx != -1:
            tmp_start_time = time[idx]
            tmp_baseline = int(row['Time']) - 5
    else:
        # find ending
        if end in event:
            idx = event.index(end)
            tmp_end_time = time[idx]
            
            duration = (tmp_baseline, tmp_start_time, tmp_end_time)
            if types == 1:
                train_neg_list.append(duration)
            elif types == 2:
                train_pos_list.append(duration)
            elif types == 3:
                test_neg_list.append(duration)
            else:
                test_pos_list.append(duration)
            
            # reset everything
            tmp_start_time = 0
            tmp_start_time = 0
            types = 0
```

```python
print(len(train_neg_list))
print(len(train_pos_list))
print(len(test_neg_list))
print(len(test_pos_list))
```

```python
# casting feature values to float, type and valence to string

feature_cols = list(data.iloc[:,0:132])# Creates list of all column headers
data[feature_cols] = data[feature_cols].astype(float)

target_cols = ['Type','Valence']
data[target_cols] = data[target_cols].astype(str)
```

```python
# labeling based on (start, end) tuples, and substracting from baseline

# labeling train negatives
for baseline,start,end in train_neg_list:
    baseline = float(baseline)
    start = float(start)
    end = float(end)
    
    baseline_row = data.loc[data['Time'] == baseline, ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])]
    
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), 'Type'] = 'Training'
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), 'Valence'] = 'Negative'
    # substracting all the rows with the baseline
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])] = data.loc[((data['Time'] > start) & (data['Time.1'] < end)), ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])].sub(baseline_row.squeeze())
    
# labeling train positive
for baseline,start,end in train_pos_list:
    baseline = float(baseline)
    start = float(start)
    end = float(end)
    
    baseline_row = data.loc[data['Time'] == baseline, ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])]
    
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), 'Type'] = 'Training'
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), 'Valence'] = 'Positive'
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])] = data.loc[((data['Time'] > start) & (data['Time.1'] < end)), ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])].sub(baseline_row.squeeze())
    
# labeling test negative
for baseline,start,end in test_neg_list:
    baseline = float(baseline)
    start = float(start)
    end = float(end)
    
    baseline_row = data.loc[data['Time'] == baseline, ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])]
    
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), 'Type'] = 'Testing'
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), 'Valence'] = 'Negative'
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])] = data.loc[((data['Time'] > start) & (data['Time.1'] < end)), ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])].sub(baseline_row.squeeze())
    
# labeling test positive
for baseline,start,end in test_pos_list:
    baseline = float(baseline)
    start = float(start)
    end = float(end)
    
    baseline_row = data.loc[data['Time'] == baseline, ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])]
    
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), 'Type'] = 'Testing'
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), 'Valence'] = 'Positive'
    data.loc[((data['Time'] > start) & (data['Time.1'] < end)), ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])] = data.loc[((data['Time'] > start) & (data['Time.1'] < end)), ~data.columns.isin(['Type', 'Valence', 'Time', 'Time.1', 'Event', 'Event.1', 'Event.2'])].sub(baseline_row.squeeze())
```
